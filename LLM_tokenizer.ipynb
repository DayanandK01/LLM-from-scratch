```python
# read the "the-verdict" book from the storage

with open("the-verdict.txt", 'r', encoding='utf-8') as f:
  raw_text = f.read()

print("Total number of character: ", len(raw_text))
print(raw_text[:99])
```

```python
# example to split the raw text read from the book into individual words for preparing tokens

import re

text = 'hello, world. this, is a test.'
result = re.split(r'(\s)', text)
print(result)
```

```python
# this is also a part of the example

result = re.split(r'([,.]|\s)', text)
print(result)
```

```python
# we should note that when we split the large sequence of text into invidual words even the white spaces
# are considered as elements of the list so we need to remove them from our list of words

result = [item for item in result if item.strip()]
print(result)
```

```python
# tokenizing our raw_text
#1. split the raw_text into individual word
#2. then strip the white spaces

preProcessedText = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text)
preProcessedText = [item.strip() for item in preProcessedText if item.strip()]
print(preProcessedText[:99])
```

```python
# converting tokens(words) into tokenId
all_words = sorted(set(preProcessedText))
vocab_size = len(all_words)
print(vocab_size)
print(all_words)
```

```python
# create a vocabulaty for the all_words (which are unique words in the preProcessedText)
# vocabulary consists of unique words as key and a unique numerical value of the values
# just like a python dictionay

vocabulary = {token: integer for integer, token in enumerate(all_words)}
```

```python
# printing first 50 vocabulary
# now we have assigned unique numerical values to unique tokens(words)
# and by doing this we have encoded the tokens into numerical values

for i, item in enumerate(vocabulary.items()):
  print(item)
  if i >= 50:
    break
```

```python
# creating a Tokenizer class with encode and decode method

class SimpleTokenizerv1:
  def __init__(self, vocabulary):
    self.str_to_int = vocabulary
    self.int_to_str = {value: key for key, value in vocabulary.items()}

  def encode(self, text):
    preProcessedText = re.split(r'([,.:;?_!"()\']|--|\s)', text)
    preProcessedText = [item.strip() for item in preProcessedText if item.strip()]
    print(f"Before assigning tokenId: \n {preProcessedText}")
    ids = [self.str_to_int[s] for s in preProcessedText]
    return ids

  def decode(self, ids):
    text = " ".join([self.int_to_str[i] for i in ids])
    print(f"Before removing the spaces infront of punctuations: \n {text}")
    # replace spaces before the specified punctuations
    text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
    return text
```

```python
# instantiating tokenizer from the SimpleTokenizer class

tokenizer = SimpleTokenizerv1(vocabulary)
# this text is present in our training set
text = """
        It's the last he painted, you know,"
        Mrs. Gisburn said with pardonable pride.
        """

ids = tokenizer.encode(text)
print(f"After assigning tokenId: \n {ids}")
```

```python
text = tokenizer.decode(ids)
print(f"Ater removing the spaces infront of punctuations: \n {text}")
```

```python
# creating a Tokenizer class with encode and decode method

class SimpleTokenizerv2:
  def __init__(self, vocabulary):
    self.str_to_int = vocabulary
    self.int_to_str = {value: key for key, value in vocabulary.items()}

  def encode(self, text):
    preProcessedText = re.split(r'([,.:;?_!"()\']|--|\s)', text)
    preProcessedText = [item.strip() for item in preProcessedText if item.strip()]
    preProcessedText = [
        item if item in self.str_to_int
        else "<|unk|>"
        for item in preProcessedText
    ]
    print(f"Before assigning tokenId: \n {preProcessedText}")
    ids = [self.str_to_int[s] for s in preProcessedText]
    return ids

  def decode(self, ids):
    text = " ".join([self.int_to_str[i] for i in ids])
    print(f"Before removing the spaces infront of punctuations: \n {text}")
    # replace spaces before the specified punctuations
    text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
    return text
```

```python
tokenizev2 = SimpleTokenizerv2(vocabulary)

text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."
text = "<|endoftext|>".join((text1, text2))
print(text)
ids = tokenizev2.encode(text)
print(f"After assigning tokenId: \n {ids}")
```

```python
text = tokenizev2.decode(ids)
print(f"Ater removing the spaces infront of punctuations: \n {text}")
```

```python
 import torch
 from torch.utils.data import Dataset, DataLoader
```

```python
class GPTDatasetv1(Dataset):

  # creating input and target tensors by looping through our entire dataset

  def __init__(self, txt, tokenizerTT, max_length, stride):
    self.input_ids = []
    self.target_ids = []

    token_ids = tokenizerTT.encode(txt, allowed_special={"<|endoftext|>"})

    for i in range(0, len(token_ids) - max_length, stride):
      input_chunk = token_ids[i: i + max_length]
      target_chunk = token_ids[i+1: i + max_length+1]
      self.input_ids.append(torch.tensor(input_chunk))
      self.target_ids.append(torch.tensor(target_chunk))

  def __len__(self):
    return len(self.input_ids)

  # returns input and target tensors of the given index
  def __getitem__(self, idx):
    return self.input_ids[idx], self.target_ids[idx]
```

```python
def create_DataLoaderv1(txt, batch_size=4, max_length=256,
                        stride=128, shuffle=True, drop_last=True,
                        num_workers=0):

  # initialize the tokenizer
  tokenizerNew = tiktoken.get_encoding("gpt2")

  # create the dataset
  dataset = GPTDatasetv1(txt, tokenizerNew, max_length, stride)

  # create the DataLoader
  dataloader = DataLoader(dataset,
                          batch_size=batch_size,
                          shuffle=shuffle,
                          drop_last=drop_last,
                          num_workers=num_workers)
  return dataloader


```
